```python
collapse_list = ['astro-ph',
'cond-mat',
'cs',
'econ',
'eess',
'gr',
'hep',
'math',
'nlin',
'nucl',
'physics',
'q-bio',
'q-fin',
'quant-ph',
'stat']
def map_string_by_case(input_string, patterns):
    # Define the patterns to match
    # patterns = ['abc', 'def', 'ghi']

    # Iterate over the patterns and check if the input string matches any of them
    for pattern in patterns:
        if input_string.lower().startswith(pattern.lower()):
            return pattern

    # If no pattern matches, return the input string
    return input_string

def collapse_categories(cat):
    return map_string_by_case(cat, collapse_list)

print(collapse_categories('stat.ML'))
```

### Basic utils

```python
# Basic utils for dates and stuff

def extract_decimal_year_of_pub(paper_versions):
    v1_date_str = next((v['created'] for v in paper_versions if v['version'] == 'v1'), None)

    if v1_date_str is None:
        return None

    v1_date = datetime.strptime(v1_date_str, '%a, %d %b %Y %H:%M:%S %Z')
    year_start = datetime(v1_date.year, 1, 1)
    year_end = datetime(v1_date.year + 1, 1, 1)

    days_since_year_start = (v1_date - year_start).days
    days_in_year = (year_end - year_start).days

    decimal_year = v1_date.year + days_since_year_start / days_in_year

    return decimal_year

def bin_date(date_in_years, bin_size):
    return math.floor(date_in_years / bin_size) * bin_size + bin_size/2.0
```

### Extract data by publication

```python
# Create a pandas dataframe with each publication information 

MAX_EXTRACTION_LENGTH = 5000000  # How many publications to load to save time 
MAX_PUBLICATIONS = 50 # max publications allowed to consider an author (otherwise it's likely a non-unique name)
MIN_PUBLICATIONS = 2

# Create an empty list to accumulate rows
rows = []

length = 0
metadata = get_metadata()
for paper in metadata:
    if length > MAX_EXTRACTION_LENGTH:
        break
    paper_data = json.loads(paper)
    
    # Extract and process the data
    paper_id = paper_data['id']
    categories = [collapse_categories(cat) for cat in paper_data['categories'].split()]
    pub_date = extract_decimal_year_of_pub(paper_data['versions'])
    authors_parsed = paper_data['authors_parsed']
    
    # Create a dictionary with the extracted data
    row = {'id': paper_id, 'categories': categories, 'pub_date': pub_date, 'authors_parsed': authors_parsed}

    # Append the row to the rows list
    rows.append(row)
    
    if length % 100_000 == 0:
        print(length, 'done')
    length += 1

# Create a DataFrame using the accumulated rows
columns = ['id', 'categories', 'pub_date', 'authors_parsed']
df = pd.DataFrame(rows, columns=columns)

print(df)
```

```python

# Get a list of all unique categories
unique_categories = sorted(set(cat for categories in df['categories'] for cat in categories))

# Function to map categories to a unit vector embedding
def category_to_vector(categories):
    vec = np.zeros(len(unique_categories))
    indices = [unique_categories.index(cat) for cat in categories]
    vec[indices] = 1.0 / len(categories)
    return vec

def index_to_category(ind):
    return unique_categories[ind]

def category_to_index(cat):
    return unique_categories.index(cat)

def index_to_verbose_category(ind):
    cat = index_to_category(ind)
    if cat in category_map:
        return cat + " : " + category_map[cat]
    else:
        return cat
    
category_names = [index_to_category(which_cat) for which_cat in range(len(unique_categories))]
category_names_verbose = [index_to_verbose_category(which_cat) for which_cat in range(len(unique_categories))]
print(unique_categories)
```

### Extract data by author

```python
# Create a dictionary with the contributions of each unique author name. 

# Initialize an empty dictionary to store the authors and their papers
author_papers = {}

# Iterate through the rows of the DataFrame
for index, row in df.iterrows():
    
    paper_id = row['id']
    categories = row['categories']
    pub_date = row['pub_date']
    authors_parsed = row['authors_parsed']

    # Iterate through the authors of each paper
    for author in authors_parsed:
        author_key = "".join(author)  # Create a unique author key

        # If the author is not in the dictionary, add them with their first paper
        if author_key not in author_papers:
            author_papers[author_key] = []

        # Add the paper information to the author's list of papers
        author_papers[author_key].append([paper_id, pub_date, categories])

print(len(author_papers))
```

### Extract derivative data by date 

```python
# Process data for each author, getting their derivatives over time. 
# Return an array of binned papers and their derivative from previous publications by the same author

YEARS_BINSIZE = 0.5  # how large bins to group together on derivative calculation

def compute_time_series_derivative(time_series):
    """
    Takes the current embedding minus the previous, and returns that at each publication time.
    Will only be nonzero if author changes fields
    """
    time_series_derivative = [(time_series[0][0], np.zeros(len(time_series[0][1])))]

    for i in range(1, len(time_series)):
        current_embedding = time_series[i][1]
        previous_embedding = time_series[i-1][1]
        difference = current_embedding - previous_embedding
        time_series_derivative.append((time_series[i][0], difference))

    return time_series_derivative

# Initialize the paper_metadata dictionary
paper_metadata = defaultdict(list)

for author_key, papers in author_papers.items():
    if not (MIN_PUBLICATIONS <= len(papers) <= MAX_PUBLICATIONS):
        continue  # only use authors with a few publications, or not too many (otherwise likely a non-unique name)

    time_series = [(p[1], category_to_vector(p[2])) for p in papers]
    time_series = sorted(time_series, key=lambda x: x[0])

    time_series_derivative = compute_time_series_derivative(time_series)

    # Store the data from each time_series and time_series_derivative
    for (date_raw, emb), (_, emb_derivative) in zip(time_series, time_series_derivative):
        date = bin_date(date_raw, YEARS_BINSIZE)
        paper_metadata[date].append((emb, emb_derivative))

# Convert lists of tuples to numpy arrays
for date, emb_list in paper_metadata.items():
    paper_metadata[date] = np.stack([np.stack((emb, emb_derivative), axis=0) for emb, emb_derivative in emb_list], axis=0)
    # Now of shape (num papers, 2, emb_size) with index1 = 0 being emb, and index1=1 being emb_derivative 
```

```python
# process data over bins

sorted_dates = list(sorted(paper_metadata.keys()))
avg_pub_emb = {date : np.average(paper_metadata[date][:,0,:], axis=0) for date in sorted_dates}
tot_pub_emb = {date : np.sum(paper_metadata[date][:,0,:], axis=0) for date in sorted_dates}
avg_deriv_emb = {date : np.average(paper_metadata[date][:,1,:], axis=0) for date in sorted_dates}
tot_deriv_emb = {date : np.sum(paper_metadata[date][:,1,:], axis=0) for date in sorted_dates}

# print(avg_deriv_emb[sorted_dates[0]][1])
```

```python
def get_weighted_transition_matrix(emb_deriv_samples):
    # The second-to-last index corresponds to the source index, and the last index corresponds to the target index.
    
    emb_size = emb_deriv_samples.shape[1]
    emb_pos = np.maximum(emb_deriv_samples, 0)
    emb_neg = np.minimum(emb_deriv_samples, 0)

    eps = 1e-10
    sum_neg = np.sum(np.abs(emb_neg), axis=1)[:, np.newaxis] + eps
    sel_neg = np.abs(emb_neg) + eps

    weights = sel_neg / sum_neg
    weighted_transitions = np.einsum("ij,ik->ijk", weights, emb_pos)
    avg_weighted_transition = np.sum(weighted_transitions, axis=0)

    print('done')
    return avg_weighted_transition

transition_metadata = {date: get_weighted_transition_matrix(paper_metadata[date][:, 1, :]) for date in sorted_dates}
```
